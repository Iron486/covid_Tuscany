```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
import pandas as pd
DATA=pd.read_csv(r"C:\Users\diego\OneDrive\Desktop\DS\kaggle_datasets\covid_Tuscany\covidars20220615.csv",sep='|')
DATA.drop(['totale_casi_positivi_ab','letalita'],axis=1,inplace=True)
#
DATA
```

```python
DATA.isna().sum()
```

```python
regione=DATA.where(DATA['geografia']=='Toscana')
regione.dropna(axis=0,how='all',inplace=True)

regione.loc[:,['deceduti_inc','totale_ricoveri_inc','tamponi_inc']]=regione.loc[:,['deceduti_inc',
        'totale_ricoveri_inc','tamponi_inc']].fillna(0)
regione.loc[:,['giorno','totale_casi_positivi',
              'deceduti','totale_ricoveri','attualmente_positivi',
              'dimessi','tamponi']]=regione.loc[:,['giorno','totale_casi_positivi',
              'deceduti','totale_ricoveri','attualmente_positivi',
              'dimessi','tamponi']].astype('int64')
regione
```

```python
regione.isna().sum()
```

```python
DATA_province=DATA.where(DATA['geografia'] != 'Toscana')
DATA_province.dropna(axis=0,how='all',inplace=True)


DATA_province.drop(['totale_ricoveri',
                        'totale_ricoveri_inc','attualmente_positivi',
                   'dimessi','tamponi','tamponi_inc'],axis=1,inplace=True)

DATA_province.loc[:,['deceduti','deceduti_inc']]=DATA_province.loc[:,['deceduti','deceduti_inc']].fillna(0)
DATA_province.loc[:,['giorno','totale_casi_positivi',
              'deceduti']]=DATA_province.loc[:,['giorno','totale_casi_positivi',
              'deceduti']].astype('int64')
DATA_province
```
